```python
## Missing Data Exercise
import pandas as pd
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
## # replace NA with 0 and using in-place parameter
car_sales_missing["Odometer"].fillna(0, inplace=True)
car_sales_missing
```

```python
# Column from Series
seats_column = pd.Series([5,5,5,5,5])
# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Column from a Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 9.6, 8.7, 9.6, 8.7, 4.0]
car_sales["fuel per 100 KM"] = fuel_economy
car_sales
```

```python
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel per 100 KM"]
car_sales 
```

```python
car_sales = car_sales.rename(columns={'Fuel per 100 KM': 'Fuel per 100 KM (L)'})
```
